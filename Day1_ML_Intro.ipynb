# What is Machine Learning?
Machine Learning (ML) is a branch of artificial intelligence that enables computers to learn from data and make decisions without being explicitly programmed. It is widely used in various fields, including cybersecurity, finance, healthcare, and more.

---



# Visualizing Machine Learning
---

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label='ML Learning Pattern', color='blue')
plt.xlabel("Input Data")
plt.ylabel("Prediction")
plt.title("Machine Learning Learning Curve")
plt.legend()
plt.show()
```

---
# How does ML differ from traditional programming?
| Traditional Programming | Machine Learning |
|-------------------------|------------------|
| Requires explicit rules and logic | Learns patterns from data |
| Manually coded decision-making | Uses statistical models to make predictions |
| Limited adaptability | Can improve with more data |
---


# Visualization: Traditional vs. Machine Learning
---

```python
categories = ['Traditional Programming', 'Machine Learning']
values = [2, 8]  # ML adapts better

plt.bar(categories, values, color=['red', 'green'])
plt.title("Traditional Programming vs Machine Learning")
plt.ylabel("Adaptability Score")
plt.show()
```

---
# Main Types of Machine Learning
1. **Supervised Learning** - Uses labeled data to learn patterns and make predictions.
2. **Unsupervised Learning** - Works with unlabeled data to find hidden structures.
3. **Reinforcement Learning** - Learns by interacting with an environment and receiving rewards.
---


# Visualizing Types of ML
---

```python
labels = ['Supervised', 'Unsupervised', 'Reinforcement']
sizes = [50, 30, 20]
colors = ['blue', 'orange', 'green']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Machine Learning Types Distribution")
plt.show()
```

# Common ML Applications
- **Fraud Detection** - Identifies fraudulent transactions in banking.
- **Recommendation Systems** - Suggests products, movies, or music based on user behavior.
- **Speech Recognition** - Converts spoken language into text.
---


# Visualization: ML Applications Popularity
---

```python
apps = ['Fraud Detection', 'Recommendation', 'Speech Recognition', 'Healthcare', 'Finance']
popularity = [80, 90, 85, 70, 75]

plt.barh(apps, popularity, color='purple')
plt.xlabel("Popularity (%)")
plt.ylabel("Applications")
plt.title("Common Machine Learning Applications")
plt.show()
```
