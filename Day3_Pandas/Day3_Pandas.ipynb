#### Pandas Series Data Structure
A **Pandas Series** is a ****one-dimensional*** labeled array capable of holding data of any type (integer, string, float, etc.). The axis labels are collectively called **indexes**.

#### Pandas DataFrame
**Pandas DataFrame** is a ***two-dimensional*** data structure with labeled **axes (rows and columns)**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```


### <a href="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv">DataSet</a>

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
# to return the first n rows, default value = 5
df.head()
# where n = 10
#df.head(10)
```

```python
#To prints information about a DataFrame (df) including the index dtype and columns, non-null values and memory usage
df.info
```

```python
#Generate statistical summary
df.describe()
```

```python
#Detect missing values
df.isnull()
```

```python
#Let us see how many are the missing values in each columns
missing_values = df.isnull().sum()
missing_values
```

```python
#Fill the missing values with median
df['Age'].fillna(df['Age'].median(), inplace=True)
#To check if the missing values were filled
df['Age'].isnull().sum()
```

```python
#Let us drop the rows where embarked column has missing values
df.dropna(subset=['Embarked'], inplace=True)
df.isnull().sum()
```

```python
# let us see the survival rate
survival_rate = df['Survived'].value_counts(normalize=True) * 100
print(f"Survival Rate:{survival_rate}")
```

```python
#Group data in class and survival
class_survival = df.groupby(['Pclass'])['Survived'].mean()*100
class_survival
```

```python
#Visualizing Survival Rate by Class
plt.figure(figsize=(8,5))
class_survival.plot(kind='bar', color=['blue','red','green'])
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
#how many female passengers in first class
female_1st_class = df[(df['Sex']=='female') & (df['Pclass']==1)]
female_1st_class.head()
```
